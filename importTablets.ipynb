```python
import csv, re, requests
from urllib import parse
```

```python
# read the data from local file
file = "data/tablets.csv"
with open(file, 'r', encoding='utf-8') as data:
    reader = csv.reader(data)
    datalist = list(reader)
```

```python
for row in datalist[1:]:
    images = row[20].split(" ")
    temp_tab,_ = Tablet.objects.get_or_create(text_reference=row[0], title=row[0])
    temp_tab.save()
    temp_reg,_ = Region.objects.get_or_create(name=row[2])
    temp_reg.save()
    temp_arch,_ = Archive.objects.get_or_create(name=row[3])
    temp_arch.save()
    temp_dos,_ = Dossier.objects.get_or_create(name=row[4])
    temp_dos.save()
    temp_place,_ = Place.objects.get_or_create(name=row[8])
    temp_place.save()
    print(temp_place.name)
    temp_scribe,_ = Scribe.objects.get_or_create(name=row[10])
    temp_scribe.save()
    temp_period,_ = Period.objects.get_or_create(name=row[11])
    temp_period.save()
    temp_tt,_ = TextType.objects.get_or_create(name=row[16])
    temp_tt.save()
    for x in images:
        temp_img,_ = TabletImage.objects.get_or_create(tablet=temp_tab,image="tablet_img/{}".format(parse.unquote(x)))
        temp_img.save()
    temp_tab.region = temp_reg
    temp_tab.archive = temp_arch
    temp_tab.dossier = temp_dos
    temp_tab.nabucco_no = row[6]
    temp_tab.museum_no = row[7]
    temp_tab.place = temp_place
    temp_tab.cdli_no = row[5]
    temp_tab.place_information = row[9]
    temp_tab.scribe = temp_scribe
    temp_tab.period = temp_period
    if row[12] == "":
        pass
    else:
        temp_tab.date_not_after = row[12]
    if row[13] == "":
        pass
    else:
        temp_tab.date_not_before = row[13]
    temp_tab.babyloneian_time = row[14]
    temp_tab.ductus = row[15]
    temp_tab.text_type = temp_tt
    temp_tab.content = row[17]
    temp_tab.bibliography = re.sub("|", "", row[19])    
    try:
        temp_tab.save()
        #print("Good: ",temp_tab)
    except:
        print("Bad: ",temp_tab)
    
    
```
